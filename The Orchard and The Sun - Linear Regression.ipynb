### Download the data

```python
# Load the dataset
# Replace 'your_dataset_path.xlsx' with the path to your dataset file
file_path = r'C:\Users\SIVA\Desktop\ML\apple_orchard_data.xlsx'
data = pd.read_excel(file_path)
```

### Feature Engineering

```python
# Define features and target variable
X = data[['Sunlight Hours']]  # Features
y = data['Number of Apples']   # Target
```

### Define and Build the Model

```python
# Create and train the linear regression model
model = LinearRegression()
model.fit(X, y)
```

### Visualize the Results

```python
# Step 6: Make predictions and visualize the results
predictions = model.predict(X)
plt.scatter(X, y, color='blue')  # Actual data
plt.plot(X, predictions, color='red')  # Predicted data
plt.title('Linear Regression: Sunlight Hours vs Number of Apples')
plt.xlabel('Sunlight Hours')
plt.ylabel('Number of Apples')
plt.show()
```

### Interpret the Results

```python
#Calculate and print accuracy metrics, intercept, and slope
mse = mean_squared_error(y, predictions)
r2 = r2_score(y, predictions)
print(f'Mean Squared Error (MSE): {mse}')
print(f'R-Squared (R2): {r2}')
print(f'Intercept: {model.intercept_}')
print(f'Slope: {model.coef_[0]}')
```
