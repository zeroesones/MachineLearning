### Pitchford Pioneers Baseball

### Install Libraries

```python
# First, install required libraries (uncomment in a local environment)
# !pip install pandas numpy matplotlib scikit-learn openpyxl
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
```

### Download the data

```python
# Load the data
df = pd.read_excel('Pitchford_Pioneers_Baseball_Data.xlsx')
```

### Feature Engineering

```python
# Data Exploration
print(df.head())
print(df.describe())

# Feature Engineering (if necessary)
# For this dataset, let's assume the features are engineered well-enough for K-means.

# Normalize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['On_Base_Percentage', 'Slugging_Percentage', 'Defensive_Plays', 'Running_Speed']])
```

### Define and Build the Model

```python
# Define the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model on the data
kmeans.fit(scaled_features)

# Predict clusters
predictions = kmeans.predict(scaled_features)
```

### Visualize the Results

```python
# Visualization
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']
for i in range(3):
    plt.scatter(scaled_features[predictions == i][:,0], scaled_features[predictions == i][:,1], c=colors[i], label=f'Cluster {i}')
plt.title('K-means Clustering')
plt.xlabel('Normalized On Base Percentage')
plt.ylabel('Normalized Slugging Percentage')
plt.legend()
plt.show()
```

### Interpret the Results

```python
# Evaluate the model
# Silhouette Score
score = silhouette_score(scaled_features, predictions)
print(f'Silhouette Score: {score}')
```

### How to improve the score?

To improve the accuracy of a K-means clustering model to at least 80%, we need to consider how K-means works and the nature of the dataset. K-means clustering is an unsupervised learning algorithm that groups data into clusters based on feature similarities. The 'accuracy' in the context of K-means is not straightforward like in supervised learning models, but we can aim for better cluster separability and coherence.

We can enhance the dataset by making the clusters more distinct. This might involve adjusting the features to create more clearly defined groups. Here are some steps we could take:

Adjust the range of values for each feature to create more distinct clusters.
Introduce some level of separation between the values of different clusters.
Reduce the overlap between data points of different clusters.

```python
# First, install required libraries (uncomment in a local environment)
# !pip install pandas numpy matplotlib scikit-learn openpyxl

import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Load the data
df = pd.read_excel('Updated_Pitchford_Pioneers_Baseball_Data.xlsx')

# Data Exploration
print(df.head())
print(df.describe())

# Feature Engineering (if necessary)
# For this dataset, let's assume the features are engineered well-enough for K-means.

# Normalize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(df[['On_Base_Percentage', 'Slugging_Percentage', 'Defensive_Plays', 'Running_Speed']])

# Define the KMeans model
kmeans = KMeans(n_clusters=3, random_state=42)

# Fit the model on the data
kmeans.fit(scaled_features)

# Predict clusters
predictions = kmeans.predict(scaled_features)

# Visualization
plt.figure(figsize=(8, 6))
colors = ['r', 'g', 'b']
for i in range(3):
    plt.scatter(scaled_features[predictions == i][:,0], scaled_features[predictions == i][:,1], c=colors[i], label=f'Cluster {i}')
plt.title('K-means Clustering')
plt.xlabel('Normalized On Base Percentage')
plt.ylabel('Normalized Slugging Percentage')
plt.legend()
plt.show()

# Evaluate the model
# Silhouette Score
score = silhouette_score(scaled_features, predictions)
print(f'Silhouette Score: {score}')
```
